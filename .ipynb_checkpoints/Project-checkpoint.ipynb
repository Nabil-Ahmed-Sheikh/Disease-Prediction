```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import f1_score, accuracy_score, confusion_matrix
import seaborn as sns
```

```python
# https://www.kaggle.com/datasets/itachi9604/disease-symptom-description-dataset/data
```

```python
df = pd.read_csv('dataset.csv')
print(df.head())
#df.describe()
df1 = pd.read_csv('Symptom-severity.csv')
print(df1.head())
```

```python
sns.set(style="whitegrid")

# Plotting using seaborn barplot
plt.figure(figsize=(15, 8))
sns.barplot(x='Symptom', y='weight', hue='Symptom', data=df1, palette='viridis')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.title('Symptom Weight Visualization')
plt.xlabel('Symptom')
plt.ylabel('Weight')
plt.show()
```

Cleaning Data

```python
df.isna().sum()
df.isnull().sum()

cols = df.columns
data = df[cols].values.flatten()

s = pd.Series(data)
s = s.str.strip()
s = s.values.reshape(df.shape)

df = pd.DataFrame(s, columns=df.columns)

df = df.fillna(0)
df.head()
```

Encoding the the symptoms with their severity weight

```python
vals = df.values
symptoms = df1['Symptom'].unique()

for i in range(len(symptoms)):
    vals[vals == symptoms[i]] = df1[df1['Symptom'] == symptoms[i]]['weight'].values[0]
d = pd.DataFrame(vals, columns=cols)
```

```python
non_integervalues = []
for row in d.values: 
    for cell in row[1:]:
        if(isinstance(cell, str) and not cell in non_integervalues):
            non_integervalues.append(cell)
            
print(non_integervalues)
```

```python


d = d.replace('dischromic _patches', 0)
d = d.replace('spotting_ urination',0)
df = d.replace('foul_smell_of urine',0)
df.head()
```

Storing the diseases and encoded symptoms in seperate dataframes

```python
(df[cols] == 0).all()

df['Disease'].value_counts()

df['Disease'].unique()

data = df.iloc[:,1:].values
labels = df['Disease'].values
```

Splitting the data

```python
X_train, X_test, y_train, y_test = train_test_split(data, labels, shuffle=True, train_size = 0.85)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression

# Initialize pipeline
pipeline = Pipeline([
    ("model", LogisticRegression(max_iter=1000))
])

from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

# Parameter grid
param_grid = [
    {
        'model': [LogisticRegression(max_iter=1000)],
        'model__C': [0.001, 0.01, 0.1, 1, 10, 100],
        'model__penalty': ['l2']  # Use only 'l2' penalty
    },
    {
        'model': [KNeighborsClassifier()],
        'model__n_neighbors': [3, 5, 7, 9],
        'model__weights': ['uniform', 'distance']
    },
    {
        'model': [SVC()],
        'model__C': [0.1, 1, 10, 100],
        'model__kernel': ['linear', 'rbf'],
        'model__gamma': ['scale', 'auto']
    }
]
```

```python
grid_search = GridSearchCV(pipeline, param_grid, cv=5, verbose=2, n_jobs=-1)
grid_search.fit(X_train, y_train)
```

```python
print("Best parameters: ", grid_search.best_params_)
print("Best cross-validation train score: ", grid_search.best_score_)
print("Best cross-validation test score: ", grid_search.score(X_test, y_test))
print("Test set accuracy: ", grid_search.best_estimator_.score(X_test, y_test))
```

```python
preds = grid_search.best_estimator_.predict(X_test)
conf_mat = confusion_matrix(y_test, preds)
df_cm = pd.DataFrame(conf_mat, index=df['Disease'].unique(), columns=df['Disease'].unique())
sns.heatmap(df_cm)
```
